## Insight #4 Analysis - Song Duration

Next I wanted to look at the average duration of songs in each year and see if there was any correlation between average duration and becoming a hit

After a quick review of the data I decided to change the song duration data into minutes from miliseconds as I thought this would make for a simpler analysis

```python
#convert song duration to minutes from milliseconds for better analysis
def time_convert(ms):
    sec=ms/1000
    return f"{int(sec//60)}:{int(sec%60)}"
durations = Spotify_analysis[['duration_ms','year']].groupby('year').mean().reset_index().iloc[0:20]
durations['min:sec'] = durations['duration_ms'].apply(time_convert)
durations
```

I then plotted the change in the duration of songs by year on a line plot to have a look at the pattern of average song duration by year!

```python
x=Spotify_analysis["year"].unique()
x=sorted(x)
y=(durations["min:sec"])
x=pd.Series(x)
plt.figure(figsize=(12,8))
sns.lineplot(x,y,data=Spotify_analysis)
plt.title('The Change in Song Duration',color = 'red',fontsize=21)
plt.xlabel('Years',size=12)
plt.ylabel('Time',size=12)
plt.show()
```

From the graph above it is clear that there is a clear, negative correlations between song duration and years. This shows me that song duration has been decreasing over the years during this time period. 

Longest and Shortest hits:

```python
# Song Lengths 
song_len = Spotify_analysis[['artist','song','duration_ms','year']]
max_dur = Spotify_analysis['duration_ms'].max()
sl_max = song_len[Spotify_analysis['duration_ms']==max_dur].T
print("Longest Duration:",sl_max)
print("\n")
min_dur = Spotify_analysis['duration_ms'].min()
sl_min = song_len[Spotify_analysis['duration_ms']==min_dur].T
print("Shortest Duration:",sl_min)
mean_dur = Spotify_analysis['duration_ms'].mean()
```

```python
#Average Durations
print("Average Duration in dataset:",time_convert(mean_dur))
mean_dur_tens = Spotify_analysis.query("Period == 'The Tens'")['duration_ms'].mean()
print("Average Duration in the Tens:",time_convert(mean_dur_tens))
mean_dur_noughties = Spotify_analysis.query("Period == 'The Noughties'")['duration_ms'].mean()
print("Average Duration in the Noughties:",time_convert(mean_dur_noughties))
```

#### From my analysis I have understood that the average song duration has been decreasing over the years for Spotify hits during the period of 2000 - 2019! The longest hit song is Mirrors by Justin Timberlake.  The shortest hit song is Old Town Road by Lil Nas X. The average duration of a top-hit song is 3mins and 48sec. The average duration of a top-hit song from 'The Tens' is 3mins and 38sec. The average duration of a top-hit song from 'The Noughties' is 3mins and 59sec.
